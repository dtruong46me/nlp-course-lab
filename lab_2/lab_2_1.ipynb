## 0. Import thư viện cần thiết

```python
import string
import re
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
```

## 1. Đọc dữ liệu trong file `wiki.txt`

```python
def read_txt_file(filepath: str) -> str:
    """
    Reads a text file and returns its content as a string.
    Args:
        filepath (str): The path to the text file.
    Returns:
        str: The content of the text file.
    """
    f = open(filepath, "r")
    content = f.readlines()
    f.close()
    return " ".join(content)

FILEPATH = "data/wiki.txt"
raw_text = read_txt_file(FILEPATH)
raw_text
```

## 2. Xây dựng hàm xử lý văn bản
- Loại bỏ ký tự số, chỉ giữ lại ký tự A-Z, a-z, `(),!?`'`
- Chuyển tất cả về chữ thường
- Loại bỏ dấu chấm câu
- Tách từ
- Loại bỏ từ dừng (stop words)
- Đem các từ về dạng gốc (stemming)
- Hàm này nhận vào một câu và trả về một danh sách các từ đã được xử lý

```python
def preprocess_text(text: str) -> str:
    """
    Preprocesses the input text by performing the following steps:
    1. Removes numeric characters, keeping only A-Z, a-z, and `(),!?`'`
    2. Converts all characters to lowercase
    3. Removes punctuation
    4. Tokenizes the text into words
    5. Removes stop words
    6. Lemmatizes the words to their base form

    Args:
        text (str): The input text to be processed.

    Returns:
        str: The cleaned and processed text.
    """
    # Step 1: Remove numeric characters, keeping only A-Z, a-z, and `(),!?`'`
    text = re.sub(pattern=r"\d+", repl="", string=text)
    text = re.sub(pattern=r"[^A-Za-z(),!?\'\`]", repl=" ", string=text)
    
    # Step 2: Convert all characters to lowercase
    text = text.lower()
    
    # Step 3: Remove punctuation
    translator = str.maketrans('', '', string.punctuation)
    text = text.translate(translator)

    # Step 4: Tokenize the text into words
    words = word_tokenize(text=text)
    
    # Step 5: Remove stop words
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    
    # Step 6: Lemmatize the words to their base form
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    
    # Step 7: Join words back into a single string
    cleaned_text = " ".join(words)

    return cleaned_text

cleaned_text = preprocess_text(text=raw_text)
cleaned_text
```
